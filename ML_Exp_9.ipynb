```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/optdigits/optdigits.tra"
data = pd.read_csv(url, header=None)
```

```python

X = data.iloc[:, :-1]
y = data.iloc[:, -1]
```

```python

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

```

```python

df_pca = pd.DataFrame(data=X_pca, columns=['PC1', 'PC2'])
df_pca['Digit'] = y
```

```python

plt.figure(figsize=(10, 7))
sns.scatterplot(x='PC1', y='PC2', hue='Digit', palette='Set1', data=df_pca, legend='full')
plt.title('PCA of UCI Optical Digits Dataset')
plt.show()

print(f'Explained variance ratio: {pca.explained_variance_ratio_}')
```
